Step 4: Data Visualization


```python
#Comparison of Internal Storage and RAM:
# Assuming 'data' is your DataFrame
# Create a scatter plot to compare internal storage and RAM
plt.figure(figsize=(8, 6))
plt.scatter(data['Internal_Storage'], data['RAM_Storage'], alpha=0.5, color='green')
plt.xlabel('Internal Storage (GB)')
plt.ylabel('RAM_Storage(GB)')
plt.title('Comparison of Internal Storage and RAM')
plt.show()
```

```python
# Distribution of Phones by Price Range:
    
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
# Create a histogram to show the distribution of phones by price range
plt.figure(figsize=(8, 6))
plt.hist(data['Price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.title('Distribution of Phones by Price Range')
plt.show()
```

```python
#Operating System Distribution:

# Assuming 'data' is your DataFrame
# Create a count plot to visualize the distribution of operating systems
plt.figure(figsize=(8, 6))
sns.countplot(x='Operating_System_Type', data=data)
plt.xticks(rotation=45)
plt.xlabel('Operating_System_Type')
plt.ylabel('Total_Mobile')
plt.title('Operating System Distribution')
plt.show()
```

```python
# Find the most common values for 'Brands', 'USB_Type', and 'Selfie_Camera'
most_common_brand = data['Brands'].value_counts().idxmax()
most_common_usb_type = data['USB_Type'].value_counts().idxmax()
most_common_selfie_camera = data['Selfie_Camera'].value_counts().idxmax()

# Display the most common values
print(f"Most Common Brand: {most_common_brand}")
print(f"Most Common USB Type: {most_common_usb_type}")
print(f"Most Common Selfie Camera: {most_common_selfie_camera}")
```

```python
import matplotlib.pyplot as plt

# Calculate value counts for each column
brand_counts = data['Brands'].value_counts()
usb_type_counts = data['USB_Type'].value_counts()
selfie_camera_counts = data['Selfie_Camera'].value_counts()

# Extract the most common value for each column
most_common_brand = brand_counts.idxmax()
most_common_usb_type = usb_type_counts.idxmax()
most_common_selfie_camera = selfie_camera_counts.idxmax()

# Create bar plots for each most common value
plt.figure(figsize=(10, 8))

# Most Common Brand
plt.subplot(3, 1, 1)
brand_counts.plot(kind='bar', color='skyblue')
plt.title(f'Most Common Brand: {most_common_brand}')
plt.xlabel('Brand')
plt.ylabel('Count')

# Most Common USB Type
plt.subplot(3, 1, 2)
usb_type_counts.plot(kind='bar', color='salmon')
plt.title(f'Most Common USB Type: {most_common_usb_type}')
plt.xlabel('USB Type')
plt.ylabel('Count')

# Most Common Selfie Camera
plt.subplot(3, 1, 3)
selfie_camera_counts.plot(kind='bar', color='lightgreen')
plt.title(f'Most Common Selfie Camera: {most_common_selfie_camera}')
plt.xlabel('Selfie Camera')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
